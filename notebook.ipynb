### Cleaning Data

```python
# Import library pandas
import pandas as pd

# Load the CSV dataset
file_path = '/customers_dataset.csv'  # Pastikan path ke file CSV benar
customers_data = pd.read_csv(file_path)

# Data Cleaning Steps

# 1. Menghapus baris duplikat berdasarkan kolom 'customer_unique_id'
customers_data_cleaned = customers_data.drop_duplicates(subset='customer_unique_id').copy()

# 2. Mengubah format teks pada kolom 'customer_city' menjadi lowercase dan 'customer_state' menjadi uppercase
customers_data_cleaned.loc[:, 'customer_city'] = customers_data_cleaned['customer_city'].str.lower()
customers_data_cleaned.loc[:, 'customer_state'] = customers_data_cleaned['customer_state'].str.upper()

# 3. Menghapus baris yang memiliki nilai hilang (missing values)
customers_data_cleaned = customers_data_cleaned.dropna()

# 4. Menghapus baris dengan kode pos (customer_zip_code_prefix) tidak valid (misalnya nilai negatif atau kosong)
customers_data_cleaned = customers_data_cleaned[customers_data_cleaned['customer_zip_code_prefix'] > 0]

# 5. Reset index setelah pembersihan
customers_data_cleaned.reset_index(drop=True, inplace=True)

# Display the cleaned dataset
print("Cleaned Data:")
print(customers_data_cleaned.head())

# Display the shape of the cleaned dataset
print("\nShape of the Cleaned Dataset:", customers_data_cleaned.shape)

# Optionally, save the cleaned data to a new CSV file
customers_data_cleaned.to_csv('cleaned_customers_data.csv', index=False)
```

**Insight:**
- Geographical Concentration: Most customers are from S達o Paulo (41,746 entries), followed by Rio de Janeiro (12,852 entries).
- Urban Customer Base: Top cities include S達o Paulo, Rio de Janeiro, and Belo Horizonte, indicating a focus on major metropolitan areas.







## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
cleaned_file_path = 'cleaned_customers_data.csv'  # Ganti dengan path ke file CSV Anda
customers_data_cleaned = pd.read_csv(cleaned_file_path)

# Set style untuk visualisasi
sns.set(style="whitegrid")

# 1. Statistik Deskriptif
print("Descriptive Statistics:")
print(customers_data_cleaned.describe())

# 2. Visualisasi distribusi customer_state
plt.figure(figsize=(12, 6))
sns.countplot(data=customers_data_cleaned, x='customer_state', order=customers_data_cleaned['customer_state'].value_counts().index)
plt.title('Distribution of Customers by State')
plt.xlabel('Customer State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()

# 3. Visualisasi top 10 kota
top_cities = customers_data_cleaned['customer_city'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='viridis')
plt.title('Top 10 Cities by Number of Customers')
plt.xlabel('Customer City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()

# 4. Visualisasi kode pos
plt.figure(figsize=(12, 6))
sns.histplot(customers_data_cleaned['customer_zip_code_prefix'], bins=30, kde=True)
plt.title('Distribution of Customer Zip Code Prefix')
plt.xlabel('Zip Code Prefix')
plt.ylabel('Frequency')
plt.show()
```

**Insight:**
### Insights:
1. **State Distribution**: Most customers are from S達o Paulo.
2. **Urban Concentration**: Top cities are S達o Paulo, Rio de Janeiro, and Belo Horizonte.

## Visualization & Explanatory Analysis

###State mana yang memiliki jumlah customer paling sedikit?:

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
cleaned_file_path = 'cleaned_customers_data.csv'  # Ganti dengan path ke file CSV Anda
customers_data_cleaned = pd.read_csv(cleaned_file_path)

# 1. Jumlah pelanggan di setiap state
state_counts = customers_data_cleaned['customer_state'].value_counts()

# Visualisasi: Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, palette='viridis')
plt.title('Number of Customers by State')
plt.xlabel('Customer State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()

# Menemukan state dengan jumlah customer paling sedikit
least_customers_state = state_counts.idxmin()
least_customers_count = state_counts.min()
print(f"The state with the least customers is {least_customers_state} with {least_customers_count} customers.")
```

### Berapa jumlah customer di setiap state?

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
cleaned_file_path = 'cleaned_customers_data.csv'  # Ganti dengan path ke file CSV Anda
customers_data_cleaned = pd.read_csv(cleaned_file_path)

# 2. Jumlah pelanggan di setiap state
state_counts = customers_data_cleaned['customer_state'].value_counts()

# Visualisasi: Pie Chart untuk distribusi pelanggan
plt.figure(figsize=(10, 10))
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Customer Distribution by State')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```

**Insight:**
1. State with Fewest Customers: Acre (AC) has the fewest customers.
2. Customer Count by State: The number of customers varies significantly, with S達o Paulo having the highest count.
